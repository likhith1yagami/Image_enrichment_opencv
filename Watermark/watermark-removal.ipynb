```
import cv2
import numpy as np

img = cv2.imread("/content/watermark-5.jpg") #define the path of the image here

alpha = 2.0
beta = -160

new = alpha * img + beta
new = np.clip(new, 0, 255).astype(np.uint8)

cv2.imwrite("/content/watermark-5.jpg", new) #overwriting the same image with the new image and watermark removed
```

```
from google.colab.patches import cv2_imshow
cv2_imshow(new)
```
